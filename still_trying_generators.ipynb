```
import os
import zipfile
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
import matplotlib.pyplot as plt

local_zip = '/tmp/cats_and_dogs_filtered.zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall('/tmp')
zip_ref.close()
file_list = [f for f in os.listdir('/tmp/cats_and_dogs_filtered/train')]
```

```
main_path = r'/tmp/cats_and_dogs_filtered/'
train_path = r'/tmp/cats_and_dogs_filtered/train'
save_path = os.path.join(main_path, 'test')
os.mkdir(save_path)

image_gen = ImageDataGenerator(rescale=1./255)

image_generator = image_gen.flow_from_directory(train_path,
                                               batch_size = 20,
                                               target_size = (150, 150),
                                               class_mode = 'binary',
                                               save_to_dir = save_path)
```

```
for x, y in image_generator:
  for i in range(len(x)):
    print(i)
  break
```
